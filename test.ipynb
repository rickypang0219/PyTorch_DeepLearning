```python
import matplotlib.pyplot as plt

def plot_order_book(order_book):
    buy_prices = [order[0] for order in order_book['buy']]
    buy_volumes = [order[1] for order in order_book['buy']]
    
    sell_prices = [order[0] for order in order_book['sell']]
    sell_volumes = [order[1] for order in order_book['sell']]

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(buy_prices, buy_volumes, width=0.4, align='center', color='green', label='Buy Orders')
    ax.bar(sell_prices, sell_volumes, width=0.4, align='edge', color='red', label='Sell Orders')

    ax.set_xlabel('Price')
    ax.set_ylabel('Volume')
    ax.set_title('Order Book')
    ax.legend()

    plt.show()

# Example order book data
order_book = {
    'buy': [
        [9.6, 2],
        [9.7, 3],
        [9.9, 1]
    ],
    'sell': [
        [10.3, 3],
        [10.4, 1],
        [10.5, 2]
    ]
}

plot_order_book(order_book)
```

```python
buy_prices = [order[0] for order in order_book['buy']]
sell_prices = [order[0] for order in order_book['sell']]

highest_buy_price = max(buy_prices)
lowest_sell_price = min(sell_prices)

reference_price = (highest_buy_price + lowest_sell_price) / 2

print("Reference Price (z):", reference_price)
```
