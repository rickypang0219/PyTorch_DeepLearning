```python
import torch 
import torch.nn as nn 
import torch.optim as optim 
import torch.functional as F 
import matplotlib.pyplot as plt 
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms
```

```python
# Hyperparams 
batch_size = 100 
input_dim = 28 * 28 # image pixel size 
output_dim = 10 # 0 - 9 numbers 
learning_rate  = 1E-2
num_epochs = 10
torch.manual_seed(2023) # set the random state
image_width = 28 
blur_size = 5
```

```python
# def transform and load data 
transform = transforms.Compose(
    [
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,)),
    ]
)


# MNIST  contains 60000 images of digits 0 - 9 
train_dataset = torchvision.datasets.MNIST(
    root="../data", train=True, transform=transform, download=True
)
```

```python
image = next(iter(train_dataset))[0]
```

```python
blur_img = [ ]
for i in range(len(train_dataset)):
    image, _ = train_dataset[i]
    height, width = image.shape[-2:]
    x = torch.randint(0, height - 5, size=(1,))
    y = torch.randint(0, width - 5, size=(1,))
    image[:, x:x+5, y:y+5].fill_(0)
    blur_img.append(image)
```

```python
for i in range(10):
    blur = blur_img[i]
    plt.imshow(torch.squeeze(blur).cpu())
    plt.show()
```
