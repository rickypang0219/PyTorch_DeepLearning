```python
import torch 
import torch.nn as nn 
import torch.optim as optim 
import torch.nn.functional as F 
from torch.distributions import MultivariateNormal
import matplotlib.pyplot as plt 
```

```python
# Generate different sets of  Gaussian data 
num_sample = 1000
dist_1 = MultivariateNormal(torch.tensor([1.,1.]), torch.tensor([[1.,0.],[0.,1.]]))
x_1 = dist_1.sample((num_sample,))

dist_2 = MultivariateNormal(torch.tensor([3.,5.]), torch.tensor([[2.,1.],[1.,2.]]))
x_2 = dist_2.sample((num_sample,))

dist_3 = MultivariateNormal(torch.tensor([6.,3.]), torch.tensor([[1.,-.5],[-0.5,1.]]))
x_3 = dist_3.sample((num_sample,))


```

```python
plt.title(" Pseudo Dataset visualization (with Labelling)")
plt.scatter(x_1[:,0], x_1[:,1], label='x_1')
plt.scatter(x_2[:,0], x_2[:,1], label='x_2')
plt.scatter(x_3[:,0], x_3[:,1],label='x_3')
plt.legend()
```

```python
# Dataset 
X = torch.cat((x_1,x_2,x_3), dim=0)
```

```python
# Concat the x_1,x_2, and x_3 together to make a 'real' data set
plt.title(r'Real dataset (without labels)')
plt.scatter(X[:,0], X[:,1])
plt.xlabel(r"feature 1")
plt.ylabel(r"feature 2")
```
